```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
dataset = pd.read_excel("EncodedData.xlsx")
#print(dataset.head)
dataset.dropna(inplace=True)
dataset.drop(columns=['Unnamed: 0'],inplace=True)
print(dataset.shape)
```

```python
data=dataset[['Requirement Stability','Requirement Change Frequency','Deadline Crossing Tendency','Deadline Crossing Frequency','Budget Extension']]
cor=data.corr(method='spearman')
# setting the parameter values
annot = True
  
# plotting the heatmap
hm = sn.heatmap(data=cor,
                annot=annot)
  
# displaying the plotted heatmap
plt.show()
```
